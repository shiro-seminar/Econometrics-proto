# Demonstration

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define x range
x=np.arange(0,10,0.1)
```

```python
#define some function
def some_function(x):
    y=np.power(x,2)+0.2
    return y
```

```python
#get value of y
y=some_function(x)
```

# Plot

```python
plt.plot(x,y)
```
